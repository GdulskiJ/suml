```python
import pandas as pd

# =========================
# Wczytanie plików
# =========================
df_svr = pd.read_csv("forecasts_5dni_all_cities_lstm.csv", parse_dates=["date"])
df_prophet = pd.read_csv("forecasts_5dni_all_cities_prophet.csv", parse_dates=["date"])

# =========================
# Ujednolicenie nazw
# =========================
df_svr = df_svr.rename(columns={"temp_forecast": "temp_svr"})
df_prophet = df_prophet.rename(columns={"temp_forecast": "temp_prophet"})

# =========================
# Merge po city + date
# =========================
df_merged = df_svr.merge(
    df_prophet[["city", "date", "temp_prophet"]],
    on=["city", "date"],
    how="inner"
)

# =========================
# Średnia z prognoz
# =========================
df_merged["temp_avg"] = (
    df_merged[["temp_svr", "temp_prophet"]].mean(axis=1)
)

# =========================
# Wynik końcowy
# =========================
df_final = df_merged[["city", "date", "temp_avg"]].copy()
df_final["temp_avg"] = df_final["temp_avg"].round(2)

# =========================
# Zapis do CSV
# =========================
df_final.to_csv("forecast_avg_5dni.csv", index=False)

print(" Zapisano plik: forecast_avg_5dni.csv")
print(df_final)
```
