```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model

# 1Ô∏è‚É£ Wczytanie modelu
model = load_model("lstm_weather_best.h5", compile=False)

# 2Ô∏è‚É£ Wczytanie czystych danych historycznych
df = pd.read_csv("european_capitals_history_clean.csv")

# 3Ô∏è‚É£ Wyb√≥r miasta
city_name = "Oslo"
city_data = df[df['city'] == city_name].sort_values('date')

# 4Ô∏è‚É£ Przygotowanie ostatnich 14 dni jako wej≈õcie do modelu
weather_cols = ['tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres']
X_input = city_data[weather_cols].values[-14:]  # ostatnie 14 dni
X_input = X_input.reshape(1, 14, 6)  # shape: (1, 14, 6)

# 5Ô∏è‚É£ Predykcja 7 dni
y_pred = model.predict(X_input)
y_pred = y_pred.reshape(7, 6)  # shape: (7 dni, 6 cech)

# 6Ô∏è‚É£ Tworzenie DataFrame z prognozƒÖ
future_dates = pd.date_range(
    start=pd.to_datetime(city_data['date'].iloc[-1]) + pd.Timedelta(days=1),
    periods=7,
    freq='D'
)

df_forecast = pd.DataFrame(y_pred, columns=weather_cols)
df_forecast['date'] = future_dates
df_forecast['city'] = city_name

print("üìÖ Prognoza 7-dniowa dla miasta:", city_name)
print(df_forecast)
```
