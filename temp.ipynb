```python
from datetime import date
import matplotlib.pyplot as plt
import meteostat as ms
import pandas as pd

# =========================
# Lokalizacja: Warszawa
# =========================
POINT = ms.Point(52.23, 21.01)  # latitude, longitude, optional elevation

# =========================
# Zakres dat
# =========================
START = date(2026, 1, 6)
END   = date(2026, 1, 14)

# =========================
# Wyszukanie pobliskich stacji (limit do 4)
# =========================
stations = ms.stations.nearby(POINT, limit=4)

# =========================
# Pobranie dziennych danych z tych stacji
# =========================
ts = ms.daily(stations, START, END)

# =========================
# Interpolacja do konkretnego punktu
# =========================
df = ms.interpolate(ts, POINT).fetch()

# =========================
# Sprawdzenie dostępnych kolumn
print(df.columns)

# =========================
# Skupienie się na średniej temperaturze (TEMP)
tavg = df['tavg']  # w nowym API jest 'temp' jako średnia temperatura

# =========================
# Wyświetlenie wyników
print(tavg)

# =========================
# Wykres średniej temperatury
plt.figure(figsize=(10,5))
plt.plot(tavg.index, tavg, marker='o', linestyle='-', color='blue')
plt.title('Średnia temperatura dzienna — Warszawa')
plt.xlabel('Data')
plt.ylabel('Temperatura (°C)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
